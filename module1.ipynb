```python
import pandas as pd
import re
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

df = pd.read_csv(url, header = None) #headless dataset
```

```python
# df.tail(10)
df.head(10) #Show the first 10 rows
```

```python
# wget https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.names
file_path = "imports-85.names"
headers = list()
with open(file_path, "r") as f:
    lines = f.readlines()

    # for line in lines:
    #     indices1 = [i for i, line in enumerate(lines) if line.startswith("7.")]
    #     indices2 = [i for i, line in enumerate(lines) if line.startswith("8.")]
    # print(indices1) => 57
    # print(indices2) => 90

    # text = "".join(lines[60:90])
    # print(text)
    
    for line in lines[60:90]:
        pattern = r"[\w-]+:" # regex =  more than 1 words & "-" with ":"
        matches = re.findall(pattern, line.strip())
        for match in matches:
            headers.append(match[0:len(match)-1])

# Expected headers
print(headers)
```

```python
df.columns = headers
# df.head(5)
```

```python
# save the adjusted dataset to a particular path
path = "/home/phuongbui/ibm_courses/ibm_da_python" + "/automobile.csv"
df.to_csv(path)
```
